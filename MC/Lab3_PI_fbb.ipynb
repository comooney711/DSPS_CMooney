```
def square(l, ax):
  ax.plot([-l, l], [-l, -l], 'k-')
  ax.plot([-l, l], [l, l], 'k-')
  ax.plot([-l, -l], [-l, l], 'k-')
  ax.plot([l, l], [-l, l], 'k-')
  
```

```
def circle(l, ax):
  theta = np.linspace(0, 360, 100)
  theta *= u.degree
  x = l * np.cos(theta.to(u.rad))
  y = l * np.sin(theta.to(u.rad))
  ax.plot(x, y, 'k-')
  
```

```
def randplot(N, l):
  x=np.random.rand(N)*2*l - l
  y=np.random.rand(N)*2*l - l
  ax.plot(x,y,'o')
```

```
fig, ax = pl.subplots(figsize=(5, 5))
square(1, ax=ax)
circle(1, ax=ax)
randplot(1000, 1)
```

```
def isin(x, y, l):
  if ((x*x + y*y) < l**2):
    return 1
  else:
    return 0
```
